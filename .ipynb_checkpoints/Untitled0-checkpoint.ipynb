```python
import seaborn as sns
from scipy.stats import bernoulli,binom,poisson,geom,uniform,norm
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_bernoulli(prob,trials=10000):
    x=np.arange(0,2)
    P_bernoulli=bernoulli.rvs(p=prob,size=trials)
    fig, ax = plt.subplots(figsize=(10,5))
    ax=sns.distplot(P_bernoulli,
                    hist=False,
                    label='Frequency Simulated')
    ax.set(xlabel='Values of X',ylabel='Frequency/Probability of X')
    ax.plot(x, bernoulli.pmf(x,prob), 'ro', label='Actual bernoulli distribution')
    ax.vlines(x, 0, bernoulli.pmf(x,prob), colors='r', lw=5, alpha=0.5)
    plt.legend()
    plt.show()
```

```python
plot_bernoulli(0.3)
```

```python
def plot_binom(n,prob,trials=1000):
    P_binom=binom.rvs(n=n,p=prob,size=trials)
    fig, ax = plt.subplots(figsize=(14,7))
    k=np.arange(0,n+1)
    ax=sns.distplot(P_binom,
                    hist=False,
                    label="Probability/Frequency Simulated")
    ax.set(xlabel='Number of successes',ylabel='Probability/Frequency')
    ax.plot(k, binom.pmf(k, n, prob), 'ro', label='Actual binomial distribution')
    ax.vlines(k, 0, binom.pmf(k, n, prob), colors='r', lw=5, alpha=0.5)
    plt.legend()
    plt.show()
```

```python
plot_binom(15,0.5)
```

```python
def plot_poisson(n,lamb,trials=2000):
    P_poisson=poisson.rvs(lamb,size=trials)
    fig, ax = plt.subplots(figsize=(14,7))
    k=np.arange(0,n+1)
    ax=sns.distplot(P_poisson,
                    hist=False,
                    label="Probability/Frequency Simulated")
    ax.set(xlabel='Number of successes',ylabel='Probability/Frequency')
    ax.plot(k, poisson.pmf(k, lamb), 'ro', label='Actual poisson distribution')
    ax.vlines(k, 0, poisson.pmf(k,lamb), colors='r', lw=5, alpha=0.5)
    plt.legend()
    plt.show()
plot_poisson(25,5)
```

```python
def plot_geometric(n,prob,trials=10000):
    P_geometric=geom.rvs(prob,size=trials)
    fig, ax = plt.subplots(figsize=(14,7))
    k=np.arange(0,n+1)
    ax=sns.distplot(P_geometric,
                    hist=False,
                    label="Probability/Frequency Simulated")
    ax.set(xlabel='Number of successes',ylabel='Probability/Frequency')
    ax.plot(k, geom.pmf(k,prob), 'ro', label='Actual geometric distribution')
    ax.vlines(k, 0, geom.pmf(k,prob), colors='r', lw=5, alpha=0.5)
    plt.legend()
    plt.show()
plot_geometric(10,0.3)
```

```python
def generator_data(a,b,size):
    n=(b-a)/(size-1)
    result=[]
    s=a
    while s<b:
        result.append(s)
        s+=n
    if len(result)<size:
        result.append(b)
    return result

def plot_uniform(a,b,n=700000):
    P_uniform=uniform.rvs(loc=a,scale=b,size=n)
    fig, ax = plt.subplots(figsize=(14,7))
    X=generator_data(a,b,100)
    ax=sns.distplot(P_uniform,
                    hist=False,
                    label="Probability/Frequency Simulated")
    ax.set(xlabel='Uniform',ylabel='Probability/Frequency')
    ax.plot(X, uniform.pdf(X,a,b), 'ro', label='Actual uniform distribution')
    plt.legend()
    plt.show()
plot_uniform(4,6)
```

```python
def plot_cdf_normal(mu,sigma):
    X=generator_data(mu-4*sigma,mu+4*sigma,1000)
    P_cdf_normal=norm.cdf(X,loc=mu,scale=sigma)
    fig, ax = plt.subplots(figsize=(14,7))
    ax.set(xlabel='Normal CDF',ylabel='Probability/Frequency')
    ax.plot(X, norm.cdf(X,mu,sigma), 'r-', label='Actual normal cummulative density function')
    plt.legend()
    plt.show()
    
    
plot_cdf_normal(0,1.5)
```

```python
def plot_pdf_normal(mu,sigma,trials=1000):
    X=generator_data(mu-4*sigma,mu+4*sigma,1000)
    P_pdf_normal=norm.rvs(size=trials,loc=mu,scale=sigma)
    fig, ax = plt.subplots(figsize=(14,7))
    ax=sns.distplot(P_pdf_normal,
                    hist=False,
                    label="Probability/Frequency Simulated")
    ax.set(xlabel='Normal PDF',ylabel='Probability/Frequency')
    ax.plot(X, norm.pdf(X,mu,sigma), 'r-', label='Actual probability density function')
    plt.legend()
    plt.show()
plot_pdf_normal(0,1.5)
```

```python
#Code example for statistics library
import statistics as stats
from fractions import Fraction
from decimal import Decimal
```

```python
#mean()
stats.mean([1,2,3,4,5,6])       #return an integer number
stats.mean([1.5,2.5,6.0])       #return a float number
stats.mean([Decimal('0.25'),    #return number as Decimal format
            Decimal('0.5'),
            Decimal('0.75')])
stats.mean([Fraction(1,3),      #return number as Fraction format
            Fraction(5,2),
            Fraction(2,3)])
try:
    stats.mean([])              #Raise StatisticError
except stats.StatisticsError:
    print("Raise StatisticsError!")
```

```python
#fmean() 
#always return float number
stats.fmean([Fraction(1,3),      #Python version >= 3.8
            Fraction(5,2),
            Fraction(2,3)])
stats.fmean([3.5, 4.0, 5.25])
try:
    stats.fmean([])              #Raise StatisticError
except stats.StatisticsError:
    print("Raise StatisticsError!")
```

```python
#geometric_mean() 
#Python version >= 3.8, always return float
#Results don't achieve exactly
stats.geometric_mean([15, 19, 7])
stats.geometric_mean([Fraction(1,3),      #Python version >= 3.8
                      Fraction(5,2),
                      Fraction(2,3)])
try:
    stats.geometric_mean([])              #Raise StatisticError
except stats.StatisticsError:
    print("Raise StatisticsError!")
try:
    stats.geometric_mean([5, 10, 0])        #Raise StatisticError
except stats.StatisticsError:
    print("Raise StatisticsError!")
try:
    stats.geometric_mean([5, 10, -20])      #Raise StatisticError
except stats.StatisticsError:
    print("Raise StatisticsError!")
```

```python
#harmonic_mean()
#version >= 3.6, always return float
#The current algorithm has an early-out when it encounters a zero in the input,
#Ex: stats.harmonic_mean([5,0,3]), output=0
stats.harmonic_mean([30, 30, 30])
stats.harmonic_mean([Fraction(1,3),      #Python version >= 3.8
               Fraction(5,2),
               Fraction(2,3)])
stats.harmonic_mean([ Decimal('0.25'),    
                Decimal('0.5'),
                Decimal('0.75')])
try:
    stats.harmonic_mean([])              #Raise StatisticError
except stats.StatisticsError:
    print("Raise StatisticsError!")
try:
    stats.harmonic_mean([5, 10, -20])      #Raise StatisticError
except stats.StatisticsError:
    print("Raise StatisticsError!")
```

```python
#median(), median_low(), median_high(), median_grouped(data,interval=1)
#Result is returned that has type following the type of number in input

stats.median({1,3,5})
stats.median([1,3,5,7])
stats.median([Decimal('0.25'),
              Decimal('0.5'),
              Decimal('0.4'),
              Decimal('0.75')])
stats.median([Fraction(1,3),
             Fraction(4,7),
             Fraction(5,2),
             Fraction(3,4)])

stats.median_low([1,3,5])
stats.median_low([1,6,5,7])        #Compare 2 middle numbers and return the lower number, like example will return 5

stats.median_high([1,3,5])
stats.median_high([1,6,5,7])       #Return the larger number (ex: 6)

stats.median_grouped([1, 2, 2, 3, 3, 4, 4, 4, 5, 6, 6, 7])
stats.median_grouped([1, 2, 2, 3, 3, 4, 4, 4, 5, 6, 6, 7], interval=2)

try:
    stats.median([])              #Raise StatisticError, median_low, median_high and median_grouped are the same
except stats.StatisticsError:
    print("Raise StatisticsError!")
```

```python
#mode()
stats.mode([1,2,3,4,5,6,7,8,1,1,2,1,3])
stats.mode([Decimal('0.1'),Decimal('0.2'),Decimal('0.3'),Decimal('0.1')])
stats.mode(['a','b','c','d','a','a'])

try:
    stats.mode(['a','b','c','d','a','b'])   #Raise StatisticsError when more than one mode found.
except stats.StatisticsError:
    print("Raise StatisticsError!")
```

```python
#multimode()
#Version >= 3.8
#Use to find more than one mode
stats.multimode(['a','b','c','d','a','b'])
stats.multimode([])
```

```python
#pvariance(data,mu=None), pstdev(data,mu=None)
stats.pvariance([0.0, 0.25, 0.5, 0.75, 1.0, 1.25, 1.75, 2.25])
stats.pvariance([Decimal("2.5"), Decimal("3.25"), Decimal("4.25"), Decimal("4.5"), Decimal("4.75")])
stats.pvariance([Fraction(1,3),Fraction(2,7),Fraction(5,6)])

try:
    stats.pvariance([])              #Raise StatisticError, pstdev() is the same
except stats.StatisticsError:
    print("Raise StatisticsError!")
    
stats.pstdev([0.0, 0.25, 0.5, 0.75, 1.0, 1.25, 1.75, 2.25]) #return the float square root of pvariance(), 
stats.pstdev([Decimal("2.5"), Decimal("3.25"), Decimal("4.25"), Decimal("4.5"), Decimal("4.75")])
stats.pstdev([Fraction(1,3),Fraction(2,7),Fraction(5,6)])
```

```python
#variance(data,xbar=None), stdev(data,xbar=None)
stats.variance([0.0, 0.25, 0.5, 0.75, 1.0, 1.25, 1.75, 2.25])
stats.variance([Decimal("2.5"), Decimal("3.25"), Decimal("4.25"), Decimal("4.5"), Decimal("4.75")])
stats.variance([Fraction(1,3),Fraction(2,7),Fraction(5,6)])

try:
    stats.variance([])              #Raise StatisticError, stdev() is the same
except stats.StatisticsError:
    print("Raise StatisticsError!")
    
stats.stdev([0.0, 0.25, 0.5, 0.75, 1.0, 1.25, 1.75, 2.25]) #return the float square root of pvariance(), 
stats.stdev([Decimal("2.5"), Decimal("3.25"), Decimal("4.25"), Decimal("4.5"), Decimal("4.75")])
stats.stdev([Fraction(1,3),Fraction(2,7),Fraction(5,6)])
```

```python
#quantiles(data,*,n=4,method='exclusive')
#version >=3.8
stats.quantiles([103, 19, 8, 86, 101, 101, 89, 81, 108, 92, 110,      #quartiles 
                 10, 75, 15, 103, 109, 76, 119, 189, 91, 103, 119,    #method='exclusive' use for data sampled from a population 
                 121, 75, 187, 12, 121, 111, 88, 89, 11, 16, 5,       #that can have more extreme values than found in the samples
                 106, 101, 4, 111, 74, 87, 86, 103, 13, 106, 86,
                 13, 107, 101, 81, 109, 104])

stats.quantiles([103, 19, 8, 86, 101, 101, 89, 81, 108, 92, 110,      #quartiles 
                 10, 75, 15, 103, 109, 76, 119, 189, 91, 103, 119,    #method='inclusive' used for describing population data, 
                 106, 101, 4, 111, 74, 87, 86, 103, 13, 106, 86,      #or for samples that are known to include the most extreme values from the population
                 121, 75, 187, 12, 121, 111, 88, 89, 11, 16, 5,
                 13, 107, 101, 81, 109, 104],method='inclusive')

stats.quantiles([103, 19, 8, 86, 101, 101, 89, 81, 108, 92, 110,      #deciles
                 10, 75, 15, 103, 109, 76, 119, 189, 91, 103, 119,
                 106, 101, 4, 111, 74, 87, 86, 103, 13, 106, 86,
                 121, 75, 187, 12, 121, 111, 88, 89, 11, 16, 5,
                 13, 107, 101, 81, 109, 104],n=10)

stats.quantiles([103, 19, 8, 86, 101, 101, 89, 81, 108, 92, 110,      #percentiles
                 10, 75, 15, 103, 109, 76, 119, 189, 91, 103, 119,
                 106, 101, 4, 111, 74, 87, 86, 103, 13, 106, 86,
                 121, 75, 187, 12, 121, 111, 88, 89, 11, 16, 5,
                 13, 107, 101, 81, 109, 104],n=100)

try:
    stats.quantiles([])              #Raise StatisticError when data not at least two data points.
except stats.StatisticsError:
    print("Raise StatisticsError!")
    
try:
    stats.quantiles([103, 19, 8, 86, 101, 101, 89, 81, 108, 92, 110,      #Raise StatisticError when n is not least 1.
                     10, 75, 15, 103, 109, 76, 119, 189, 91, 103, 119,
                     106, 101, 4, 111, 74, 87, 86, 103, 13, 106, 86,
                     121, 75, 187, 12, 121, 111, 88, 89, 11, 16, 5,
                     13, 107, 101, 81, 109, 104],n=0)
except stats.StatisticsError:
    print("Raise StatisticsError!")
```
